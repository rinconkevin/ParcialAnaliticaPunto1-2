Se importan las librerias para poder graficar y calcular el algoritmo.


```python
# Tratamiento de datos
# ==============================================================================
import pandas as pd
import numpy as np

# Gr√°ficos
# ==============================================================================
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
from mlxtend.plotting import plot_decision_regions

# Preprocesado y modelado
# ==============================================================================
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Configuraci√≥n matplotlib
# ==============================================================================
plt.rcParams['image.cmap'] = "bwr"
#plt.rcParams['figure.dpi'] = "100"
plt.rcParams['savefig.bbox'] = "tight"
style.use('ggplot') or plt.style.use('ggplot')

# Configuraci√≥n warnings
# ==============================================================================
import warnings
warnings.filterwarnings('ignore')
```

Luego se saca el signal 1 y 2 para los valores de "x" y "y" en la grafica, tambien se crea una lista llamada y que me indicara los puntos claves de cada latido gracias al ANN. y se forma el dataset.


```python
from pandas.io.formats.style_render import DataFrame

c1 = pepito.p_signal
c2 = pepito2.p_signal
c3 = pepito3.p_signal
signal1=c2[:,0]
signal2=c2[:,1]

ann=wfdb.rdann('16272','atr', sampfrom = 0, sampto= 100000)### etiquetas del evento de interes
print(ann)
ANN=ann.sample
print(ANN)

y= np.zeros(100000)

for i in ANN:
  y[i] = 1;
  for j in range(5):
    if((i+j)<100000 and (i-j)>0):
      y[i+j] = 1;
      y[i-j] = 1;

d = {"X1": signal1, "X2":signal2, "y": y}


Datos = pd.DataFrame(data=d)


```

Se realiza una grafica punteada o de dispersion.para visualizar cuales son los puntos con latido y cuales no. 

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.scatter(Datos.X1, Datos.X2, c=y);
ax.set_title("Datos ESL.mixture");
```

Vista general de los datos. "x", y "y" validacion del latido 1 = latido, 0 = no latido.

```python
Datos.head()
```

Se realiza el testeo con las variables de entrenamiento.

```python
# Divisi√≥n de los datos en train y test
# ==============================================================================
X = Datos.drop(columns = 'y')
y = Datos['y']

X_train, X_test, y_train, y_test = train_test_split(
                                        X,
                                        y.values.reshape(-1,1),
                                        train_size   = 0.8,
                                        random_state = 1234,
                                        shuffle      = True
                                    )
```

Se crea el modelo lineal del SVM con las variables de entrenamiento.

```python
# Creaci√≥n del modelo SVM lineal
# ==============================================================================
modelo = SVC(kernel = 'linear')
modelo.fit(X_train, y_train)
```

se cuadran algunos parametros como el minimo y el maximo de entrenamientos para luego empezar a graficar el SVM intentando ceparar los latidos de los no latidos de una forma lineal.

```python
# Representaci√≥n gr√°fica de los l√≠mites de clasificaci√≥n
# ==============================================================================
# Grid de valores
x = np.linspace(np.min(X_train.X1), np.max(X_train.X1), 50)
y = np.linspace(np.min(X_train.X2), np.max(X_train.X2), 50)
Y, X = np.meshgrid(y, x)
grid = np.vstack([X.ravel(), Y.ravel()]).T

# Predicci√≥n valores grid
pred_grid = modelo.predict(grid)

fig, ax = plt.subplots(figsize=(6,4))
ax.scatter(grid[:,0], grid[:,1], c=pred_grid, alpha = 0.2)
ax.scatter(X_train.X1, X_train.X2, c=y_train, alpha = 1)

# Vectores soporte
ax.scatter(
    modelo.support_vectors_[:, 0],
    modelo.support_vectors_[:, 1],
    s=200, linewidth=1,
    facecolors='none', edgecolors='black'
)

# Hiperplano de separaci√≥n
ax.contour(
    X,
    Y,
    modelo.decision_function(grid).reshape(X.shape),
    colors = 'k',
    levels = [-1, 0, 1],
    alpha  = 0.5,
    linestyles = ['--', '-', '--']
)

ax.set_title("Resultados clasificaci√≥n SVM lineal");
```

Luego se hace un test de las predicciones. 

```python
# Predicciones test
# ==============================================================================
predicciones = modelo.predict(X_test)
predicciones
```

Para finalmente hacer la comparacion y sacar el porcentaje de precision o desempe√±o.

```python
# Accuracy de test del modelo 
# ==============================================================================
accuracy = accuracy_score(
            y_true    = y_test,
            y_pred    = predicciones,
            normalize = True
           )
print("")
print(f"El accuracy de test es: {100*accuracy}%")
```
